```python
# import library
import matplotlib.pyplot as plt
import cv2
```

```python
#  image read
img = cv2.imread("/content/img.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Edge detection
edges = cv2.Canny(blurred, 50, 150)

# Contour detection
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Drawing contours on a copy of the original image
img_contours = img.copy()
cv2.drawContours(img_contours, contours, -1, (0, 255, 0), 2)

# Show
plt.imshow(img_contours)
plt.axis("off")
plt.title("Contours Image")
plt.show()
```

# Extracting individual objects from the image

```python
# Extracting individual objects from the image
for i, cnt in enumerate(contours):
    x, y, w, h = cv2.boundingRect(cnt)
    obj = img[y:y+h, x:x+w]
    cv2.imwrite(f"object_{i}.jpg", obj)

# Show individual objects from the image
for i, cnt in enumerate(contours):
    x, y, w, h = cv2.boundingRect(cnt)
    obj = img[y:y+h, x:x+w]

    # نمایش شیء
    plt.figure(figsize=(6, 6))  # تنظیم اندازه پنجره نمایش
    plt.imshow(cv2.cvtColor(obj, cv2.COLOR_BGR2RGB))  # تبدیل رنگ از BGR به RGB
    plt.title(f"Object {i}")
    plt.axis('off')  # حذف محورهای x و y
    plt.show()
```

# Drawing a rectangle around each object

```python
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Show
plt.imshow(img)
plt.axis("off")
plt.title("Objects")
plt.show()

```
