```python
import cv2
import matplotlib.pyplot as plt
my_img = cv2.imread("/content/image.jpg")
gray = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.axis("off")
plt.title("Gray Image")
plt.show()
```

# Applying blur filter  basic noise reduction

```python
Blur = cv2.GaussianBlur(my_img, (7, 7), 0)
plt.imshow(Blur)
plt.axis("off")
plt.title("Blur Image")
plt.show()
```

# Median Blur

```python
median = cv2.medianBlur(my_img, 5)
plt.imshow(median)
plt.axis("off")
plt.title("Median Image")
plt.show()
```

# Canny (Edge Detection)

```python
edges = cv2.Canny(my_img, 90, 180)
plt.imshow(edges)
plt.axis("off")
plt.title("Edges Image")
plt.show()
```

# Bilateral  Filter

```python
bilateral = cv2.bilateralFilter(my_img, 9, 75, 75)
plt.imshow(bilateral)
plt.axis("off")
plt.title("Bilateral Image")
plt.show()
```
