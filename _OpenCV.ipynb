# Read And Show Image

```python
import cv2
import matplotlib.pyplot as plt
# read image
my_img = cv2.imread("/content/image.jpg")
# print(my_img)
# show image
my_img_RGB = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
plt.imshow(my_img_RGB)
plt.axis("off")
plt.show()
```
