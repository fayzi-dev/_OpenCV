# Read And Show Image

```python
import cv2
import matplotlib.pyplot as plt
# read image
my_img = cv2.imread("/content/image.jpg")
# print(my_img)
# show image
my_img_RGB = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
plt.imshow(my_img_RGB)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

# Image color space transformation and storage

```python
gray = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")
plt.show()
```

```python
img_HSV = cv2.cvtColor(my_img, cv2.COLOR_BGR2HSV)
plt.imshow(gray)
plt.title("HSV Image")
plt.axis("off")
plt.show()
```
